```python
# ==========================================
# Imports and Setup
# ==========================================
# %load_ext autoreload
# %autoreload 2

!pip install mediapy # Install mediapy if not already installed
import physical_ai_av
import numpy as np
import mediapy as mp
from physical_ai_av.utils import tf
import matplotlib.pyplot as plt

# Initialize interface
avdi = physical_ai_av.PhysicalAIAVDatasetInterface()

# ==========================================
# Fetch Data
# ==========================================
clip_id = "99a9ff74-d5cb-4f58-8364-bdd32e611c4f"
camera = avdi.features.CAMERA.CAMERA_FRONT_TELE_30FOV
prediction_timestamp_us = 5_100_000

# Get video data
video = avdi.get_clip_feature(clip_id, feature=camera, maybe_stream=True)

frame_timestamps_us = prediction_timestamp_us + 100_000 * np.arange(-10, 1)
frames, actual_timestamps_us = video.decode_images_from_timestamps(frame_timestamps_us)

# Get calibration data
extrinsics = avdi.get_clip_feature(
    clip_id, feature=avdi.features.CALIBRATION.SENSOR_EXTRINSICS, maybe_stream=True
)
intrinsics = avdi.get_clip_feature(
    clip_id, feature=avdi.features.CALIBRATION.CAMERA_INTRINSICS, maybe_stream=True
)
dimensions = avdi.get_clip_feature(
    clip_id, feature=avdi.features.CALIBRATION.VEHICLE_DIMENSIONS, maybe_stream=True
)

# Get egomotion data
egomotion = avdi.get_clip_feature(clip_id, feature=avdi.features.LABELS.EGOMOTION, maybe_stream=True)

# Display video (Optional)
mp.show_video(frames, fps=10, width=512)


# ==========================================
# Demo of using TF Tree & Trajectory Plot
# ==========================================
# First, set up the transform tree
tf_tree = tf.TransformTree()
tf_tree.add_transform("anchor", "ego", egomotion)
for sensor_id, sensor_pose in extrinsics.sensor_poses.items():
    tf_tree.add_transform("ego", sensor_id, sensor_pose)

# Now, let's get the egomotion data in the frame of the camera at the last frame timestamp
frame_time = actual_timestamps_us[-1]
anchor_to_camera_frame = tf_tree.lookup_transform(
    tf.FrameInfo(camera, frame_time), tf.FrameInfo("anchor")
)

query_times = np.linspace(frame_time, frame_time + 5_000_000, 50)
egomotion_in_camera_frame = egomotion(query_times).transform(
    anchor_to_camera_frame.tf_target_source
)

egomotion_pose_in_camera_image = intrinsics.camera_models[camera].ray2pixel(
    egomotion_in_camera_frame.pose.translation
)

# Don't draw out of bounds pixels
egomotion_pose_in_camera_image = np.where(
    intrinsics.camera_models[camera].is_out_of_bounds(egomotion_pose_in_camera_image)[..., None],
    np.nan,
    egomotion_pose_in_camera_image
)

# Plot the image along with the transformed egomotion
fig, ax = plt.subplots(1)
ax.imshow(frames[-1])
h = ax.scatter(
    egomotion_pose_in_camera_image[:, 0],
    egomotion_pose_in_camera_image[:, 1],
    c=(query_times - frame_time) / 1e6,
    s=10,
    cmap="rainbow",
)
plt.colorbar(h, label="Lookahead Time (s)", ax=ax)
plt.show()
plt.tight_layout()


# ==========================================
# Bounding Box Projection Plot
# ==========================================
# Use the vehicle dimensions to plot the bounding box
bbox_corners_xyz_ego_frame = np.array(
    [
        [dimensions.length / 2, dimensions.width / 2, 0],
        [dimensions.length / 2, -dimensions.width / 2, 0],
        [-dimensions.length / 2, -dimensions.width / 2, 0],
        [-dimensions.length / 2, dimensions.width / 2, 0],
        [dimensions.length / 2, dimensions.width / 2, 0],
    ]
)
bbox_corners_xyz_ego_frame[..., 0] -= dimensions.rear_axle_to_bbox_center

# --- CRITICAL FIX: Bypass SciPy Shape Strictness ---
# Apply the 50 future poses to each of the 5 bounding box corners individually,
# then stack them together to safely create the (5, 50, 3) array.
poses = egomotion(query_times).pose
transformed_corners = [poses.apply(corner) for corner in bbox_corners_xyz_ego_frame]
bbox_corners_xyz_anchor_frame = np.stack(transformed_corners)

# Apply the camera transform to the flattened (250, 3) array
bbox_corners_xyz_camera_frame = anchor_to_camera_frame.tf_target_source.apply(
    bbox_corners_xyz_anchor_frame.reshape(-1, 3)
).reshape(5, -1, 3)

bbox_corners_uv_camera_image = intrinsics.camera_models[camera].ray2pixel(
    bbox_corners_xyz_camera_frame
)

# Don't draw out of bounds bounding boxes
bbox_corners_uv_camera_image = np.where(
    intrinsics.camera_models[camera]
    .is_out_of_bounds(bbox_corners_uv_camera_image)
    .any(axis=0)[None, :, None],
    np.nan,
    bbox_corners_uv_camera_image,
)

# Plot the image along with the transformed egomotion
fig, ax = plt.subplots(1)
ax.imshow(frames[-1])

# Create a colormap and normalizer for the Lookahead time
cmap = plt.cm.rainbow
norm = plt.cm.colors.Normalize(vmin=0, vmax=5)

# Plot each bounding box with color scaled by the Lookahead time
for i in range(bbox_corners_uv_camera_image.shape[1]):
    lookahead_time = (query_times[i] - frame_time) / 1e6
    color = cmap(norm(lookahead_time))
    ax.plot(
        bbox_corners_uv_camera_image[:, i, 0],
        bbox_corners_uv_camera_image[:, i, 1],
        color=color,
        linewidth=2,
    )

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, label="Lookahead Time (s)", ax=ax)
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# 1. SETUP, AUTHENTICATION & REPO CLONING
# ==========================================
!pip install -q physical_ai_av==0.2.2 huggingface_hub matplotlib numpy opencv-python-headless pillow pandas torch pyarrow fastparquet
!git clone https://github.com/NVlabs/alpamayo1.5.git

import os
import sys
import random
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from huggingface_hub import login, HfApi, hf_hub_download
from PIL import Image

# Add NVIDIA's internal src folder to Colab's Python path
sys.path.append(os.path.abspath('./alpamayo1.5/src'))

import physical_ai_av
from alpamayo1_5.load_physical_aiavdataset import load_physical_aiavdataset

# --- Authentication ---
# Replace with your token or use the interactive login
login()

# --- Configuration ---
CONFIG = {
    "pool_size_to_load": 20,
    "num_to_visualize": 10,
    "t0_us": 5_100_000
}

# ==========================================
# 2. UTILITY FUNCTIONS
# ==========================================
def stitch_all_cameras_to_image(img_tensor):
    """Extracts the latest frame from all cameras and stitches them horizontally."""
    # Shape: [num_cams, num_frames, C, H, W]
    if img_tensor.ndim == 5:
        latest_frames = img_tensor[:, -1, :, :, :]
    else:
        latest_frames = img_tensor

    imgs_np = latest_frames.permute(0, 2, 3, 1).cpu().numpy()

    # Reverse ImageNet normalization
    img_min, img_max = imgs_np.min(), imgs_np.max()
    if img_max > img_min:
        imgs_np = (imgs_np - img_min) / (img_max - img_min)

    panorama = np.concatenate(np.clip(imgs_np, 0, 1), axis=1)
    return panorama

# ==========================================
# 3. METADATA & SENSOR INVENTORY
# ==========================================
print("âŒ“ Initializing Dataset Interface...")
avdi = physical_ai_av.PhysicalAIAVDatasetInterface()

print("âŒ“ Fetching metadata to find available clips...")
available_clips = []
sensor_db = None

try:
    api = HfApi()
    repo_files = api.list_repo_files(repo_id="nvidia/PhysicalAI-Autonomous-Vehicles", repo_type="dataset")
    # Using feature_presence to identify available modalities
    target_file = next(f for f in repo_files if 'feature_presence' in f and f.endswith('.parquet'))

    metadata_path = hf_hub_download(repo_id="nvidia/PhysicalAI-Autonomous-Vehicles", repo_type="dataset", filename=target_file)
    sensor_db = pd.read_parquet(metadata_path)

    # Extract UUIDs from index as seen in grafik_16.png
    available_clips = [c for c in sensor_db.index.dropna().unique().tolist() if len(str(c)) > 10]
    print(f"  âœ… SUCCESS! Extracted {len(available_clips)} UUIDs.")
except Exception as e:
    print(f"âŒ Metadata fetch failed: {repr(e)}")

# ==========================================
# 4. DATA LOADING & PHYSICS CALCULATION
# ==========================================
valid_samples = []
if available_clips:
    random.shuffle(available_clips)
    print(f"\nðŸš€ Downloading pool of {CONFIG['pool_size_to_load']} clips...")

    for clip_id in available_clips:
        if len(valid_samples) >= CONFIG["pool_size_to_load"]: break

        try:
            # Deep parser extracts tensors for images and ego-motion
            data = load_physical_aiavdataset(str(clip_id), t0_us=CONFIG["t0_us"], avdi=avdi)

            # --- Calculate True Physics ---
            hist_xyz = data["ego_history_xyz"].cpu().numpy().squeeze()
            hist_rot = data["ego_history_rot"].cpu().numpy().squeeze()

            speed, yaw_rate = 0.0, 0.0
            if hist_xyz.ndim >= 2 and len(hist_xyz) >= 2:
                dt = 0.1 # 10Hz sampling
                speed = float(np.linalg.norm(hist_xyz[-1] - hist_xyz[-2]) / dt)
                # Correcting the format error by ensuring yaw_rate is a float scalar
                yaw_rate = float((hist_rot[-1, 2] - hist_rot[-2, 2]) / dt)

            valid_samples.append({
                "id": str(clip_id),
                "image": data["image_frames"],
                "traj": data["ego_future_xyz"],
                "speed": speed,
                "yaw_rate": yaw_rate
            })
        except Exception: continue

# ==========================================
# 5. BEV VISUALIZATION & OVERLAY
# ==========================================
if not valid_samples:
    print("ðŸš¨ No valid samples loaded.")
else:
    selected = random.sample(valid_samples, min(CONFIG["num_to_visualize"], len(valid_samples)))
    fig = plt.figure(figsize=(24, 6 * len(selected)))
    gs = gridspec.GridSpec(len(selected), 2, width_ratios=[2.5, 1])

    for i, sample in enumerate(selected):
        # Column 1: Panorama + Telemetry
        ax_img = fig.add_subplot(gs[i, 0])
        ax_img.imshow(stitch_all_cameras_to_image(sample["image"]))
        ax_img.set_title(f"Sample {i+1} | {sample['id'][:12]}...", fontweight="bold")
        ax_img.axis("off")

        # Overlay using formatted float scalars
        overlay = f"EGO VEHICLE STATE\nSpeed: {sample['speed']:.2f} m/s\nYaw Rate: {sample['yaw_rate']:.4f} rad/s"
        ax_img.text(0.01, 0.95, overlay, transform=ax_img.transAxes, fontsize=12,
                    color='white', fontweight='bold', va='top', family='monospace',
                    bbox=dict(boxstyle='round', facecolor='black', alpha=0.6))

        # Column 2: BEV Trajectory
        ax_traj = fig.add_subplot(gs[i, 1])
        traj = sample["traj"].cpu().numpy().squeeze()
        if traj.ndim > 2: traj = traj[0]

        ax_traj.plot(-traj[:, 1], traj[:, 0], 'g-o', markersize=3, label="Future BEV")
        ax_traj.plot(0, 0, 'r^', markersize=12, label="Ego")
        ax_traj.set_xlim(-20, 20)
        ax_traj.set_ylim(-5, 45)
        ax_traj.grid(True, linestyle='--')
        ax_traj.set_title("Bird's-Eye View Trajectory")

    plt.tight_layout()
    plt.show()
```
